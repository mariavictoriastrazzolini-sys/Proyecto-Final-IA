### **Introducción de la problemática a abordar**

El aprendizaje del idioma inglés representa una dificultad frecuente para muchas personas que poseen un nivel inicial o básico. En muchos casos, los materiales educativos disponibles presentan explicaciones complejas, estructuras gramaticales avanzadas o contenidos poco adaptados al nivel del estudiante principiante.

Esta situación genera frustración y desmotivación, provocando que muchos estudiantes abandonen el proceso de aprendizaje antes de adquirir habilidades básicas como la comprensión de textos simples, la escritura de frases cortas o la correcta pronunciación de palabras cotidianas.

En un contexto donde el conocimiento del idioma inglés resulta cada vez más importante para el acceso a oportunidades académicas, laborales y personales, resulta relevante desarrollar herramientas que faciliten su aprendizaje de forma accesible y personalizada.

La inteligencia artificial generativa ofrece nuevas posibilidades para abordar esta problemática, ya que permite generar contenido educativo adaptado al nivel del estudiante de forma rápida, flexible y personalizada.

### **Solución propuesta**

La solución propuesta consiste en utilizar modelos de inteligencia artificial generativa para producir material educativo orientado a estudiantes principiantes del idioma inglés.

El sistema se basa en la creación de prompts diseñados específicamente para generar contenido educativo, como por ejemplo:

explicaciones de vocabulario básico

textos simples en inglés

ejercicios de práctica

guías de pronunciación

Para la implementación del proyecto se utilizarán dos tipos de modelos de IA:

Modelo texto-texto

Permite generar explicaciones, textos educativos, ejercicios y traducciones adaptadas al nivel inicial del estudiante.

Ejemplo de prompt:

"Explicá 5 palabras básicas en inglés relacionadas con la vida cotidiana para estudiantes principiantes hispanohablantes. Incluí la traducción al español y un ejemplo en una oración simple."

Modelo texto-imagen

Permite generar imágenes educativas que faciliten el aprendizaje visual del vocabulario.

Ejemplo de prompt:

"Illustration for beginners learning English showing basic objects labeled in English such as house, food, water, book and dog, simple educational style, colorful, clear background."

La combinación de ambos modelos permite crear una experiencia de aprendizaje más dinámica y comprensible.

```python
prompt_zero = """
Genera 5 palabras básicas en inglés para estudiantes principiantes.
Incluye:
- palabra en inglés
- traducción al español
- una oración simple en inglés.
"""

response_zero = text_model.generate_content(prompt_zero)

print("Resultado Zero-Shot:\n")
print(response_zero.text)
```

```python
prompt_one = """
Ejemplo:

Dog - Perro
Sentence: The dog is running.

Ahora genera otra palabra siguiendo el mismo formato.
"""

response_one = text_model.generate_content(prompt_one)

print("Resultado One-Shot:\n")
print(response_one.text)
```

```python
prompt_few = """
Ejemplos:

House - Casa
Sentence: The house is big.

Book - Libro
Sentence: I read a book.

Water - Agua
Sentence: I drink water.

Ahora genera 3 palabras nuevas siguiendo el mismo formato.
"""

response_few = text_model.generate_content(prompt_few)

print("Resultado Few-Shot:\n")
print(response_few.text)
```
